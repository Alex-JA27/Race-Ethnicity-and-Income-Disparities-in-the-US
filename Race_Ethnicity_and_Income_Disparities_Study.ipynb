# Project Proposal: Analyzing Income Inequality by Race and Ethnicity in the 21st Century

```python
import pandas as pd
import matplotlib.pyplot as plt

#Open the Excel file
data = pd.read_excel("Earnings_and_Height")

#Clean the file with only the data that we're going to work with
data = data[["race","earnings"]]

#Check how much data we have(check how many people per race)
data["race"].value_counts()
```

```python
# Assign a new variable with the specific data, in this case per each race. 
#If the race is equal to a certain number it will save all the rows
#with that number to the new variable

data_white = data[data["race"] == 1]
data_black = data[data["race"] == 2]
data_hispanics = data[data["race"] == 3]
data_others = data[data["race"] == 4]
```

```python
def median_formula(race_earnings):
    '''
    Calculates the median earnings from a DataFrame filtered by race.
    '''
    median_race = 0
    race_sorted_earnings = sorted(race_earnings["earnings"])
    if len(race_sorted_earnings) % 2 != 0:
        # If the length is odd, get the middle element
        median_race = race_sorted_earnings[(len(race_sorted_earnings) // 2)]
    else:
        # If the length is even, get the average of the two middle elements
        mid1 = race_sorted_earnings[(len(race_sorted_earnings) // 2 - 1)]
        mid2 = race_sorted_earnings[(len(race_sorted_earnings) // 2)]
        median_race = (mid1 + mid2) / 2
    
    return median_race
```

```python
median_white = median_formula(data_white)
median_black = median_formula(data_black)
median_hispanics = median_formula(data_hispanics)
median_others = median_formula(data_others)
```

```python
#Create a histogram of Earnings for White people
#this histogram will show how frequently people earn a certain amount of money
plt.hist(data_white["earnings"], edgecolor = "black", linewidth = 1.2)
plt.grid(axis = "y",alpha = 0.5)
plt.title("Histogram of Earnings of White People")
plt.xlabel("Earnings")
plt.ylabel("Frequency")
plt.axvline(x = median_white, color = 'black', linestyle = "dashed",label = "Median")
plt.legend()
plt.savefig("White People")
plt.show()
```

```python
#Create a histogram of Earnings for Black people
#this histogram will show how frequently people earn a certain amount of money
plt.hist(data_black["earnings"], color = "green", edgecolor = "black", linewidth = 1.2)
plt.grid(axis = "y",alpha = 0.5)
plt.title("Histogram of Earnings of Black People")
plt.xlabel("Earnings")
plt.ylabel("Frequency")
plt.axvline(x = median_black, color = 'black', linestyle = "dashed",label = "Median")
plt.legend()
plt.savefig("Black People.png")
plt.show()
```

```python
#Create a histogram of Earnings for Hispanic people
#this histogram will show how frequently people earn a certain amount of money
plt.hist(data_hispanics["earnings"], color = "red", edgecolor = "black", linewidth = 1.2, alpha = 0.8)
plt.grid(axis = "y",alpha = 0.5)
plt.title("Histogram of Earnings of Hispanics/Latin People")
plt.xlabel("Earnings")
plt.ylabel("Frequency")
plt.axvline(x = median_hispanics, color = 'black', linestyle = "dashed",label = "Median")
plt.legend()
plt.savefig("Hispanics People.png")
plt.show()
```

```python
#Create a histogram of Earnings for Other races
#this histogram will show how frequently people earn a certain amount of money
plt.hist(data_others["earnings"], color = "purple", edgecolor = "black", linewidth = 1.2, alpha = 0.8)
plt.grid(axis = "y",alpha = 0.5)
plt.title("Histogram of Earnings of Other Races")
plt.xlabel("Earnings")
plt.ylabel("Frequency")
plt.axvline(x = median_others, color = 'black', linestyle = "dashed",label = "Median")
plt.legend()
plt.savefig("Other Races .png")
plt.show()
```

```python
#Create a histogram with all the races to see the difference in Earnings for each race
#this histogram will show how frequently people earn a certain amount of money per race
plt.hist(data_white["earnings"], edgecolor = "black", linewidth = 1.2, alpha = 0.8, label = "White")
plt.hist(data_black["earnings"], color = "green", edgecolor = "black", linewidth = 1.2, label = "Black")
plt.hist(data_hispanics["earnings"], color = "red", edgecolor = "black", linewidth = 1.2, alpha = 0.8, label = "Hispanics/Latinos")
plt.hist(data_others["earnings"], color = "purple", edgecolor = "black", linewidth = 1.2, alpha = 0.8, label = "Others")
plt.grid(axis = "y",alpha = 0.5)
plt.title("Histogram of Earnings by Race/Ethnicity")
plt.xlabel("Earnings")
plt.ylabel("Frequency")
plt.savefig("Project Figure.png")
plt.legend()
plt.show()
```
